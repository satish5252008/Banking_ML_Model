```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
train=pd.read_excel('C:\\Users\\tvimal\\Desktop\\TEXT_ANALYSIS_HLB\\Cards_Transaction.xlsx',delimiter='\t',quoting=3)
train.head()
```

```python
#NUMBER OF WORDS IN TRANSACTION DESCRIPTION
train['word_count'] = train['Tran_Desc'].apply(lambda x: len(str(x).split("-")))
train[['Tran_Desc','word_count']].sort_values('word_count',ascending=False).head()
```

```python
#NUMBER OF CHARACTERS
train['char_count'] = train['Tran_Desc'].str.len() ## this also includes spaces
train[['Tran_Desc','char_count']].head()
```

```python
#Avarage word Length
def avg_word(sentence):
    words = sentence.split(" ")
    return (sum(len(word) for word in words)/len(words))

train['avg_word'] = train['Tran_Desc'].apply(lambda x: avg_word(x))

train[['Tran_Desc','avg_word']].head()
```

```python
#NUmber of special characters
train['hastags'] = train['Tran_Desc'].apply(lambda x: len([x for x in x.split() if x.startswith('-')]))
train[['Tran_Desc','hastags']].sort_values('hastags',ascending=False).head()
```

```python
#converting lower case into upper case
train['Tran_Desc'] = train['Tran_Desc'].apply(lambda x: " ".join(x.upper() for x in x.split()))
train['Tran_Desc'].head()
```

```python
train.Tran_Desc = train.Tran_Desc.str.rstrip().astype(str)
```

```python
#df.column into list
lst=list(train['Tran_Desc'])
#convert list into list of list
list_of_list=[s.split('  ') for s in lst] #using double space [[i] for i in lst]
#joining the string by single space , before that split by double space
output = [' '.join(i).split("  ") for i in list_of_list]
# To create a country column
train['Tran_Desc'] = train['Tran_Desc'].str.replace(',', '')# only one row has comma 
#splitting last three characters in each sublist
single_text = ','.join(train['Tran_Desc'].tolist()).split(',')#df1['Tran_Desc'].tolist()   --- can use this also
sample=[]
for i in range(len(single_text)):
    country=single_text[i][-3:]
    sample.append(country)
#Now converting list of lisst into dataframe by double space
s=pd.DataFrame(output)
s.replace('', np.nan, inplace=True) # df has empty cells in some places 
#splitting merchant name and location then 
place=s.iloc[:,1:]
raw=s.iloc[:,0].rename(columns={0:'merchant'})
#add all location columns into a single column
combine_place=place.apply(lambda x: ','.join(x.dropna()), axis=1)
#merging merchant name and state
fin=pd.concat([raw,combine_place],axis=1)
#change the column name 
fin=fin.rename(columns={0:'MERCHANT',1:'STATE'})
fin.head()
```

```python

#now split the state column into 2 by using comma to get country
stat= pd.DataFrame(fin.STATE.str.split(',',1).tolist(),columns = ['STATES','COUNTRY']) # 1 represent number of columns need
#now dataframe have two state columns so drop older one and concat the others to get name,state and country
new=pd.concat([fin,stat],axis=1).drop('STATE',axis=1)
#state column have numbers so replaced by empty space
new['new_state'] = new['STATES'].str.replace('\d+', '')
#it creates the new column if state column has mys
new['final_state']=new.new_state.str.contains('MYS')
#To get all country create new one and if true then put mys
new['COUNTRYS'] = new['COUNTRY']
new.loc[new['final_state'] == True, 'COUNTRYS'] = 'MYS'
#now we got country in seperate column so remove mys from sate column
new['new_state'] = new['new_state'].map(lambda x: x.lstrip('').rstrip('MYS'))
#Now  to get all country name ,check if any merchant column have mys  if yes follow same thing we did for state column
#next four lines will remove only mys from merchant name column
new['final_MERCHANT']=new.MERCHANT.str.contains('MYS')
new['NEW_MERCHANT'] = new['MERCHANT']
new.loc[new['final_MERCHANT'] == True, 'NEW_MERCHANT'] = new['MERCHANT']
new['NEW_MERCHANT'] = new['NEW_MERCHANT'].map(lambda x: x.lstrip('').rstrip('MYS'))
#To get clear merchant name  take the first string before any special characters
new['MERCHANT_NAME_FINAL']=new['NEW_MERCHANT'].str.replace("^([a-zA-Z]+(?:\s+[a-zA-Z]+)*).*", r"\1")
# sample have last three  characterss, already we created so create the final country column
new['COUNTRY_F']=sample
#take only required columns
required_text=new[['MERCHANT_NAME_FINAL','new_state','COUNTRY_F']]
#concat with uncleaned text for better understanding
FINAL_REQUIREMENT=pd.concat([train,required_text],axis=1)
FINAL_REQUIREMENT.head()
```
